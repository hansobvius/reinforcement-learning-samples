# Robot Corridor - Learning Visualization
This notebook visualizes the training logs from `data/training_log.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Load Data
# We use '..' to go up one level from 'notebooks/' to the project root
data_path = os.path.join('..', 'data', 'training_log.csv')
df = pd.read_csv(data_path)

print(df.head())
```

```python
# Plot Steps per Episode
plt.figure(figsize=(10, 5))
plt.plot(df['Episode'], df['Steps'], marker='o', linestyle='-', color='b')
plt.title('Steps to Reach Diamond per Episode')
plt.xlabel('Episode')
plt.ylabel('Steps')
plt.grid(True)
plt.show()
```

```python
# Plot Total Reward per Episode
plt.figure(figsize=(10, 5))
plt.plot(df['Episode'], df['TotalReward'], marker='o', linestyle='-', color='g')
plt.title('Total Reward per Episode')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.grid(True)
plt.show()
```
